```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
print("OpenCV-Python Version %s" %cv2.__version__)
```

```python
img=cv2.imread('messi.jpg',cv2.IMREAD_COLOR)
if img is None:
    print('Open Error')
else:
    print('Image Loaded')
```

```python
#display an image using matplotlib
#plt.imshow(img)
#the color of this line is wrong
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
```

```python
#check the resulting img
print (type(img))
print (img.shape)
print (img.dtype)
print (img[:2,:2,:])
```

```python
#create a black image
img2=np.zeros((512,512,3),np.uint8)
plt.imshow(img2)
```

```python
#draw a line using cv2.line(image,startpoint)
cv2.line(img2,(0,0),(511,511),(255,0,0),5)
#diagonal thick line with 5px

#draw a rectanle using cv2.rectangle(image,topleft,bottom)
cv2.rectangle(img2,(384,0),(510,128),(0,255,0),3)
#green rectangle with thickness of 3 px

#draw a circle using cv2.circle(image,center,radius,rgb)
cv2.circle(img2,(447,63),63,(0,0,255),-1)
#blue filled circle(thickness is -1)

#draw an eelipse using cv2.ellipse(image,center,axes,angular)
cv2.ellipse(img2,(256,256),(100,50),-45,0,180,(255,0,0),-1)
#red wide down half ellipse

plt.imshow(img2)
```
